```python
import numpy as np
import math
import matplotlib.pyplot as plt

testlist = np.array(range(-20,20))
analytical_value = np.exp(testlist)
taylor_sum = np.zeros(len(testlist), dtype=float)

#Using the original Taylor Expansion

for j in range(len(testlist)):
    for i in range(20):
        taylor_sum[j] += pow(testlist[j], i)/float(math.factorial(i))

relative_error = np.absolute(taylor_sum)/analytical_value
plot1, = plt.semilogy(testlist, relative_error, label = "Original Exponential Expansion")

#Modification: Use 1/exp(x) instead:
for i in range(len(range(-15, 0))):
    taylor_sum[i] = 1.0/taylor_sum[-i-1]
relative_error1 = taylor_sum/analytical_value

plot2, = plt.semilogy(testlist, relative_error1, label = "Modified Exponential Expansion")

plt.title("Relative Error")
plt.legend(handles = [plot1, plot2])
plt.show()
```
